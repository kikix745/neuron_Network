```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import math
```

```python
X,y=make_blobs(n_samples=100,n_features=2,random_state=0,centers=2)
y=y.reshape((y.shape[0],1))
print('dim x is:',X.shape)
print('dim y is:',y.shape)
plt.scatter(X[:,0],X[:,1],c=y,cmap="summer")
plt.show()
```

```python
#fonction d'initialisation
def initialisation(X):
    W=np.random.randn(X.shape[1],1)
    b=np.random.randn(1)
    return (W,b)

W,b=initialisation(X)
b
```

```python
# le model 
def model(X,W,b):
    Z=X.dot(W)+b
    A=1/(1+np.exp(-Z))
    return A

```

```python
# fonction de cout log loss
def log_loss(A,y):
    eps=1e-15
    L=(1/len(y))*np.sum(-y*np.log(A+eps)-(1-y)*np.log(1-A+eps))
    return L
```

```python
 #gradient 
def gradient (A,X,y):
    dW=(1/len(y))*np.dot(X.T,A-y)
    db=(1/len(y))*np.sum(A-y)
    return (dW,db)
```

```python
#update 
def update(dW,db,W,b,learning_rate):
    W=W-learning_rate*dW
    b=b-learning_rate*db
    return(W,b)
```

```python
from sklearn.metrics import accuracy_score
```

```python
#the main artificial neuron
def artificial_neuron(X,y,learning_rate=0.1,n_iter=100):
    #initialiser W,b
    W,b=initialisation(X)
    Loss=[]
    for i in range (n_iter):
        A=model(X,W,b)
        Loss.append(log_loss(A,y))
        dW,db=gradient(A,X,y)
        W,b=update(dW,db,W,b,learning_rate)

    
#plot loss function during the training
    plt.plot(Loss)
    plt.show()
    return W,b
```

```python
#showing the plot 
artificial_neuron(X,y,learning_rate=0.1,n_iter=100)
```

```python
# prediction 
def predict(X,W,b):
    A=model(X,W,b)
    return A>=0.5
y_pred= predict(X,W,b)
print(y_pred)
# accuracy of the mdel
print("the accuracy is :",accuracy_score(y,y_pred))
print("the W is : ",W,"the b is : ",b)
```
